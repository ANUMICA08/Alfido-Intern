```python
import io
import pandas as pd

# Replace with your actual file name after upload
df = pd.read_csv(io.BytesIO(uploaded['ecommerce_customer_data_custom_ratios.csv']))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
sns.set(style="whitegrid")
pd.set_option('display.max_columns', None)
```

```python
print("ðŸ”¹ Shape:", df.shape)
print("\nðŸ”¹ Data Types:\n", df.dtypes)
print("\nðŸ”¹ Missing Values:\n", df.isnull().sum())
print("\nðŸ”¹ First 5 Rows:\n", df.head())

# Remove duplicates and missing rows
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
print("\nðŸ”¹ Descriptive Statistics:")
print(df.describe())

# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 1. Gender Distribution
if 'Gender' in df.columns:
    sns.countplot(data=df, x='Gender', palette='Set2')
    plt.title("Gender Distribution")
    plt.show()

# 2. Age Distribution
if 'Age' in df.columns:
    sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
    plt.title("Age Distribution")
    plt.show()

# 3. Annual Income Distribution
income_cols = [col for col in df.columns if 'Income' in col]
for col in income_cols:
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()

# 4. Spending Score Distribution
spending_cols = [col for col in df.columns if 'Spending' in col]
for col in spending_cols:
    sns.histplot(df[col], bins=30, kde=True, color='orange')
    plt.title(f"Distribution of {col}")
    plt.show()

# 5. Boxplot of Spending Score by Gender
if 'Gender' in df.columns and spending_cols:
    sns.boxplot(x='Gender', y=spending_cols[0], data=df, palette='pastel')
    plt.title(f"{spending_cols[0]} by Gender")
    plt.show()

# 6. Age vs Spending Score
if 'Age' in df.columns and spending_cols:
    sns.scatterplot(data=df, x='Age', y=spending_cols[0], hue='Gender', palette='coolwarm')
    plt.title(f"Age vs {spending_cols[0]}")
    plt.show()
```

```python
import pandas as pd
import io

# Replace the filename with the actual name after uploading
df = pd.read_csv(io.BytesIO(uploaded['superstore_final_dataset (1) (1).csv']), encoding='latin1')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
pd.set_option('display.max_columns', None)
```

```python
# Basic info
print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nFirst 5 Rows:\n", df.head())

# Drop duplicates and missing values
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

# Convert date columns if they exist
for col in df.columns:
    if 'Date' in col or 'date' in col:
        df[col] = pd.to_datetime(df[col], dayfirst=True, errors='coerce')
```

```python
print("\nDescriptive Statistics:\n", df.describe())

# Correlation Heatmap (numerical columns only)
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 1. Customer Segmentation (if column exists)
if 'Segment' in df.columns:
    sns.countplot(x='Segment', data=df, palette='Set2')
    plt.title("Customer Segments")
    plt.show()

# 2. Sales by Category
if 'Category' in df.columns and 'Sales' in df.columns:
    category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
    category_sales.plot(kind='bar', color='lightcoral')
    plt.title("Total Sales by Category")
    plt.ylabel("Sales")
    plt.show()

# 3. Profit by Region
if 'Region' in df.columns and 'Profit' in df.columns:
    region_profit = df.groupby('Region')['Profit'].sum()
    region_profit.plot(kind='bar', color='seagreen')
    plt.title("Total Profit by Region")
    plt.ylabel("Profit")
    plt.show()

# 4. Sales vs Profit Scatter
if 'Sales' in df.columns and 'Profit' in df.columns:
    sns.scatterplot(x='Sales', y='Profit', data=df, hue='Category', alpha=0.7)
    plt.title("Sales vs Profit by Category")
    plt.show()

# 5. Monthly Trends (if Order Date exists)
if 'Order Date' in df.columns:
    df['Month'] = df['Order Date'].dt.to_period('M')
    monthly_sales = df.groupby('Month')['Sales'].sum()
    monthly_sales.plot(kind='line', marker='o')
    plt.title("Monthly Sales Trend")
    plt.ylabel("Sales")
    plt.xticks(rotation=45)
    plt.show()
```
